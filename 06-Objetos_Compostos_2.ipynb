## 1 – Um objeto composto multidimensional (matriz) pode receber  objeto composto unidimensional (vetor)?

```python
obj_1 = [1, 2, 3]
obj_2 = ["1", "2", "3"]
obj_3 = ["a", "b", "c"]

obj_composto = [
    obj_1,
    obj_2,
    obj_3
]

obj_composto
```

R.: Sim, objetos compostos multidimensionais podem receber diversos objetos compostos unidimensionais.

## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
lista_genes = ["gloC", "ascA", "p54", "mtrABC", "piliIV"]

gloC = [len(lista_genes[0]),
        lista_genes[0][0:2],
        lista_genes[0].casefold().count("a"),
        "hu" in lista_genes[0].casefold()]

ascA = [len(lista_genes[1]),
        lista_genes[1][0:2],
        lista_genes[1].casefold().count("a"),
        "hu" in lista_genes[1].casefold()]

p54 = [len(lista_genes[2]),
       lista_genes[2][0:2],
       lista_genes[2].casefold().count("a"),
       "hu" in lista_genes[2].casefold()]

mtrABC = [len(lista_genes[3]),
         lista_genes[3][0:2],
         lista_genes[3].casefold().count("a"),
         "hu" in lista_genes[3].casefold()]

piliIV = [len(lista_genes[4]),
          lista_genes[4][0:2],
          lista_genes[4].casefold().count("a"),
          "hu" in lista_genes[4].casefold()]

matriz_genes = [gloC, ascA, p54, mtrABC, piliIV]

matriz_genes

#matriz.insert(postion, object)
```

## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
data_df = pd.read_csv("./data/data.csv")


data_dict = {
    "Número de entradas": data_df.shape[0],
    "Classes das colunas": [type(column) for column in data_df.columns]
}

data_dict["Número de classe 'str'"] = [data_dict["Classes das colunas"].count(str)]

data_dict["Média"] = [data_df.iloc[:, 0].mean(),
                      data_df.iloc[:, 3].mean(),
                      data_df.iloc[:, 8].mean()]

data_dict["Desvio Padrão"] = [data_df.iloc[:, 0].std(),
                              data_df.iloc[:, 3].std(),
                              data_df.iloc[:, 8].std()]

data_dict["Terceiro Quartil"] = [data_df.iloc[:, 0].quantile(0.75),
                              data_df.iloc[:, 3].quantile(0.75),
                              data_df.iloc[:, 8].quantile(0.75)]

#data_df.info() # Informações gerais do DataFrame
#data_df.describe() # Descreve colunas quantitativas
#print(data_df.describe().iloc[2, [0, 3, 8]]) # Isolar valor descritivo
#data_df.select_dtypes(exclude="number").describe() # Descreve colunas qualitativas
data_dict
```
