## 04 – Escreva o programa em Python:

```python
def superiorADez(array):
    for i in range(len(array)):
        if (array > 10)[i] == True:
            print(f"{array[i]} é maior que 10 ({(array > 10)[i]})")
    print(f"Total maiores que 10: {(array >10).sum()}")
```

```python
array_atividade_um = np.random.normal(loc=10, scale=3, size=5)
# loc = média, scale = desvio padrão, size = quantidade de valores

superiorADez(array_atividade_um)
```

Logo, os valores gerados tendem a não se aproximarem de 20 e portando deverão não superar esse valor.

## 07 – Escreva o programa em Python:

```python
def calculaMedia(valores):
    media = (sum(valores))/len(valores)
    return media
```
