## 01 – Dados as operações abaixo, em Python, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Imprima os objetos e o resultado usando uma estrutura condicional.

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t

print(f"""
      p -> {p},
      q -> {q},
      r -> {r},
      s -> {s},
      t -> {t}
""")

if resultado == True:
    print(f"É {resultado}! :)")
elif resultado == False:
    print(f"É {resultado}! :(")
```

## Instancie um objeto da classe array numpy, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3. 
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R.: 
#### **Primeira solução**
Como o resultado de (array_atividade_um > 10) é também um array de True e False com a mesma indexação do array original, pode-se usar um loop 'for' no range() do tamanho do array para indexar cada um dos elementos no array que também corresponde ao mesmo index do operador lógico (array_atividade_um > 10). Dessa forma é possível usar um condicional para mostrar e contar apenas os valores que forem verdadeiros (>10).

#### **Segunda solução (sem condição nem looping):**
Também é possivel evitar o uso de um loop e um condicional usando a indexação do array através da máscara de booleano (array_atividade_um > 10). Isso retorna apenas os elementos do array que são positivos a máscara booleana e também é possivel contar quantos elementos são retornados nesse array filtrado (usando len()) para saber o total.

## 04 – Escreva o programa em Python:

```python
# loc = média, scale = desvio padrão, size = quantidade de valores
array_atividade_um = np.random.normal(loc=10, scale=3, size=5)

for i in range(len(array_atividade_um)):
    if (array_atividade_um > 10)[i] == True:
        print(f"{array_atividade_um[i]} é maior que 10 ({(array_atividade_um > 10)[i]})")
print(f"Total maiores que 10: {(array_atividade_um >10).sum()}")
```

```python
print(f"Valores superiores a 10: {array_atividade_um[(array_atividade_um > 10)]}")
print(f"Total de valores superiores a 10: {len(array_atividade_um[(array_atividade_um > 10)])}")
```

Logo, os valores gerados tendem a não se aproximarem de 20 e portando deverão não superar esse valor.

```python
max_value = 20

for i in range(len(array_atividade_um)):
    if (array_atividade_um > max_value)[i] == True:
        print(f"{array_atividade_um[i]} é maior que 10 ({(array_atividade_um > max_value)[i]})")
print(f"Total maiores que {max_value}: {(array_atividade_um > max_value).sum()}")
```

## 09 – Escreva o programa em Python:

```python
genes_df = pd.read_csv("./data/genes_df.csv", index_col="Unnamed: 0")
genes_df.index = ["Número de caracteres", "Duas primeiras letras", "Contagem de 'a'", "Presença de 'hu'"]

genes_df
```

```python
primeiro_gene = genes_df.columns[0]
num_caracteres = len(primeiro_gene)
print(f"Primeiro gene: {primeiro_gene}")
print(f"Número de caracteres: {num_caracteres}")

if (2 <= num_caracteres <= 4):
    print(f"O nome '{primeiro_gene}' tem entre 2 e 4 caracteres.")
    print("R.: SIM, posso saber as informações do gene")
    print(f"Informações do gene '{primeiro_gene}':")
    display(genes_df[primeiro_gene])

elif num_caracteres < 2:
    print(f"O nome '{primeiro_gene}' tem menos de 2 caracteres.")
    print("Genes com número de caracteres superior a 4:")
    genes_maior_4 = [col for col in genes_df.columns if len(col) > 4]
    display(genes_df[genes_maior_4])

else:
    print(f"O nome '{primeiro_gene}' tem mais de 4 caracteres.")
    print(f"Genes com número de caracteres superiores a '{primeiro_gene}':")
    genes_maior = [col for col in genes_df.columns if len(col) > len(primeiro_gene)]
    display(genes_df[genes_maior])
```
