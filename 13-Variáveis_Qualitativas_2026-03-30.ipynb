## Importe o .csv utilizado no exercício 7 da aula 11 e instancie um objeto da classe DataFrame. Considere que não à uma ordem entre os genes.
Como inserir  na DataFrame uma tabela de frequência do número de caracteres no nome dos genes? Uma vez inserida, como obter um gráfico de barras deste dado?

```python
genes_df = pd.read_csv("./data/genes_df.csv", index_col="Unnamed: 0")
genes_df.index = ["Número de caracteres", "Duas primeiras letras", "Contagem de 'a'", "Presença de 'hu'"]
genes_df = genes_df.T
genes_df["Número de caracteres"] = genes_df["Número de caracteres"].astype(int)

genes_df
```

## 01 – Descreva o racional da sua resposta:

R.: Como a frequência de caracteres é definida pela razão entre o número de caracteres do gene e soma de todos os caracteres entre todos os genes, podemos realizar essa operação de divisão e guardar esse valor em uma nova coluna com as frequências individuais para cada gene.

## 03 – Escreva o programa em Python:

```python
caracteres_max = genes_df["Número de caracteres"].sum()

genes_df["Frequência de Caracteres"] = (genes_df["Número de caracteres"] / caracteres_max).round(3)

genes_df
```

## Importe o .csv utilizado no exercício 7 da aula 11 e instancie um objeto da classe DataFrame. Considere que à uma ordem entre os genes.
Como inserir  na DataFrame uma tabela de frequência do número de caracteres no nome dos genes? Uma vez inserida, como obter um gráfico de barras deste dado?

```python
genes_df = pd.read_csv("./data/genes_df.csv", index_col="Unnamed: 0")
genes_df.index = ["Número de caracteres", "Duas primeiras letras", "Contagem de 'a'", "Presença de 'hu'"]
genes_df = genes_df.T
genes_df["Número de caracteres"] = genes_df["Número de caracteres"].astype(int)

genes_df
```

## 04 – Descreva o racional da sua resposta:

R.: O racional é o mesmo da resposta anterior: "Como a frequência de caracteres é definida pela razão entre o número de caracteres do gene e soma de todos os caracteres entre todos os genes, podemos realizar essa operação de divisão e guardar esse valor em uma nova coluna com as frequências individuais para cada gene".<br> No entanto, agora organizamos o dataframe por ordem alfabética para simular uma ordem fixa.

## 05 – Desenhe o fluxograma da resposta:

<img src="./img/13-frequencia.png" height="600" />

## 06 – Escreva o programa em Python:

```python
genes_df = genes_df.sort_index()

caracteres_max = genes_df["Número de caracteres"].sum()

genes_df["Frequência de Caracteres"] = (genes_df["Número de caracteres"] / caracteres_max).round(3)

genes_df
```

```python
genes_df.plot.bar(y="Frequência de Caracteres")
```

## 07 – Reorganize a DataFrame, removendo as colunas contendo os dois primeiros caracteres, o número de caracter 'a', e a presença de 'hu', dos genes. Renomeie as colunas de acordo com as informações. Adicione uma coluna com a informação de 'Amostra 1'.

```python
genes_df.drop(columns=["Contagem de 'a'", "Duas primeiras letras", "Presença de 'hu'"], inplace=True)

genes_df = genes_df.rename(columns={
    "Número de caracteres": "Caracteres",
    "Frequência de Caracteres": "Amostra 1"
})

genes_df
```

## 08 – Usando a função randint(), da classe random do módulo numpy (np.random.randint(1,10,5)), instancie objeto da classe numpy array. Calcule as frequências desta nova amostra, considerando que à uma ordem entre os genes. Insira na DataFrame este objeto como "Amostra 2".

```python
array_sete = np.random.randint(1, 10, 5)

frequencia_sete = array_sete / array_sete.sum()

genes_df["Amostra 2"] = np.round(frequencia_sete, 3)

genes_df
```

## 09 – Repita o exercício 8, inserindo na DataFrame as replicas "Amostra 3“, "Amostra 4“, "Amostra 5“ e "Amostra 6".

```python
for i in range(3, 7):
    array = np.random.randint(1, 10, 5)
    frequencia = array / array.sum()
    genes_df[f"Amostra {i}"] = np.round(frequencia, 3)

genes_df
```
