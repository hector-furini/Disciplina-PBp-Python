## 04 – Instancie uma variável da classe lista, contendo 5 nomes de genes diferentes, recupere de cada um dos valores quantas letras ‘a’ estão presentes:

```python
lista_genes = ["gloC", "ascA", "p54", "mtrABC", "piliIV"]
for gene in lista_genes:
    print(f"{gene} -> Tem {gene.casefold().count('a')} letra(s) 'a'")
```

## 09 – Instancie cinco variáveis da classe lista. Cada uma delas será referente à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Em cada uma das listas, você deve inserir o número de caracteres presentes no nome do gene, os dois primeiros caracteres do nome, quantas vezes aparece o caractere ‘a’ e se no nome aparecem os caracteres ‘hu’. Para isso você deve utilizar a indexação de lista. Retorne cada um dos objetos:

```python
gloC = [len(lista_genes[0]),
        lista_genes[0][0:2],
        lista_genes[0].casefold().count("a"),
        "hu" in lista_genes[0].casefold()]

ascA = [len(lista_genes[1]),
        lista_genes[1][0:2],
        lista_genes[1].casefold().count("a"),
        "hu" in lista_genes[1].casefold()]

p54 = [len(lista_genes[2]),
       lista_genes[2][0:2],
       lista_genes[2].casefold().count("a"),
       "hu" in lista_genes[2].casefold()]

mtrABC = [len(lista_genes[3]),
         lista_genes[3][0:2],
         lista_genes[3].casefold().count("a"),
         "hu" in lista_genes[3].casefold()]

piliIV = [len(lista_genes[4]),
          lista_genes[4][0:2],
          lista_genes[4].casefold().count("a"),
          "hu" in lista_genes[4].casefold()]

print(gloC, ascA, p54, mtrABC, piliIV, sep="\n")
```
