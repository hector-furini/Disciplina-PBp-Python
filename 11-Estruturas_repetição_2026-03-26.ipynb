## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Consigo verificar se o número é par ou impar verificando se o resto da sua divisão por 2 é 0 ou não. Dessa forma, consigo usar uma estrutura de repetição juntamente com uma condicional para contar elementos pares.

## 02 – Desenhe o fluxograma da resposta:

<img src="./img/11-random_list.png" width="400" />

## 03 – Escreva o programa em Python:

```python
lista_ex_um = list(np.random.randint(0, 10000, 5000))

lista_ex_um_pares = [num for num in lista_ex_um if num%2 == 0]
print(f"A lista tem {len(lista_ex_um)} valores.")
print(f"{len(lista_ex_um_pares)} são pares.")
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
contador = 0
lista_ex_quatro_pares = []

while contador < len(lista_ex_um):
    if lista_ex_um[contador]%2 == 0:
        lista_ex_quatro_pares.append(lista_ex_um[contador])
    contador += 1
print(f"A lista tem {len(lista_ex_um)} valores.")
print(f"{len(lista_ex_quatro_pares)} são pares.")
```

## Importe o .csv exportado na última aula e instancie um objeto da classe DataFrame.

Dos números de letras nos genes, quais são pares e quantos são impares?

```python
genes_df = pd.read_csv("./data/genes_df.csv", index_col="Unnamed: 0")
genes_df.index = ["Número de caracteres", "Duas primeiras letras", "Contagem de 'a'", "Presença de 'hu'"]
genes_df = genes_df.T
genes_df["Número de caracteres"] = genes_df["Número de caracteres"].astype(int)

genes_df
```

## 05 – Descreva o racional da sua resposta:

R.: Usando uma estrutura de repetição em cima do número de caracteres, consigo aplicar mesma a lógica do resto da divisao por 2 == 0 em cada um dos elementos.

## 06 – Desenhe o fluxograma da resposta:

<img src="./img/11-genes_df.png" width="400" />

## 07 – Escreva o programa em Python:

```python
genes_df["No. caracteres é par"] = [True if num%2 == 0 else False for num in genes_df["Número de caracteres"]]

genes_df
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
resultado = []
contador = 0

while contador < len(genes_df["Número de caracteres"]):
    num = genes_df["Número de caracteres"][contador]
    resultado.append(num % 2 == 0)
    contador += 1

genes_df["No. caracteres é par"] = resultado

genes_df
```
